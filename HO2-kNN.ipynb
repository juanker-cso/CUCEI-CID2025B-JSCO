# Clasificación por K-vecinos cercanos

Este método de clasificación es una técnica por la cual se agrupan entradas en un conjunto de datos según una cantidad arbitraria $k$ de las otras entradas más similares según su distancia en la métrica seleccionada. Por ejemplo, en la geometría euclideana en un plano cartesiano $x,y$, la distancia entre dos puntos $a=(x_1,y_1)$ y $b=(x_2,y_2)$ es calculada con:

$$
d = \sqrt{(x_2 - x_1)^{2} + (y_{2} - y_{1})^{2}}
$$

Dado un conjunto ya clasificado de datos, entonces, se considera que un nuevo dato entrará a una clasificación si entre sus $k$ vecinos hay más de esta clase que de otras.

El valor de $k$ se selecciona arbitrariamente, pero es necesario ser cuidadoso. Ya que es posible crear un modelo que este _sobre-ajustado_ a los valores de entrenamiento, y que falle cuando se implemente a entradas nuevas. Será necesario probar distintos valores de $k$ y medir el rendimiento del modelo resultante para encontrar su valor óptimo.

Por su síglas en inglés, se le conoce como kNN, o k-Nearest Neighbors.

## Modelo matemático

Como generalización, para un espacio de $d$ dimensiones con puntos $a$ y $b$ coordenados en el espacio, su distancia euclideana es:

$$
d(a,b) = \sqrt{\sum^{d}_{j} (a_{j} - b_{j})^{2}}
$$

Para encontrar los $k$ vecinos más cercanos a un punto $X_j$ en el conjunto $X$, es necesario calcular la distancia de $X_j$ a todos los demás puntos $X_{i,i\neq j}$, ordenarlos de menor a mayor, y tomar la clase de los primeros $k$ resultados. La clase con mayor frecuencia en estos resultados será la clase de $X_j$.

Si existe un empate entre 2 o más clases, se recurre a una ponderación: usualmente se asigna peso a cada vecino en función de su distancia específica.

## Scikit KNeighborsClassifier

La clase `sklearn.neighbors.KNeighborsClassifier` se utiliza para crear un modelo de clasificación kNN, al asignar el constructor, se puede inmediatamente seleccionar el valor de $k$. La clase se encarga por si sola de resolver conflictos de empate, aunque también se puede modificar el sistema de ponderación en el constructor.
