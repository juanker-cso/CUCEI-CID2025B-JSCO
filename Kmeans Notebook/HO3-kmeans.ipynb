# Clasificación por k-means

Es uno de los métodos de clasificación en aprendisaje máquina más utilizados. Funciona de manera similar al método kNN; pero, mientras que el segundo se dedica a clasificar un dato nuevo contra categorías existentes, k-means realiza la agrupación de datos no clasificados. Esto se logra de manera iterativa bajo el siguiente proceso:

1. Asigna un valor $k$ arbitrario
2. Selecciona $k$ entradas aleatoreas en el plano como centroides
3. Para cada punto se busca el centroide más cercano y se clasifican de acuerdo a este
4. Para cada grupo, se calcula un nuevo centroide como la posición promedio de los miembros
5. Se repiten los pasos 3 y 4, recalculando la clasificación de cada punto y encontrando los nuevos centroides
6. El proceso se detiene si una iteración del paso 3 tiene la misma composición que la iteración anterior

Debido a la condición inicial aleatoria del proceso, es posible que dos ejecuciones consecutivas del proceso obtengan resultados muy diferentes. Por lo que se sugiere crear varios modelos y tomar el que otoge mejores métricas de calidad.
